```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

model_id = "Bllossom/llama-3.2-Korean-Bllossom-3B"

# 1. ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì € ë¶ˆëŸ¬ì˜¤ê¸° (4ë¹„íŠ¸ ì–‘ìí™”ë¡œ ë©”ëª¨ë¦¬ ì ˆì•½)
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    low_cpu_mem_usage=True
)

# 2. íŒŒì´í”„ë¼ì¸ ì„¤ì •
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=512,
    temperature=0.7,
    top_p=0.9
)
```

```python
# 4ï¸âƒ£ ë³€í™˜ í•¨ìˆ˜ (ìµœì¢… ìˆ˜ì • ë²„ì „)
import re

def keep_only_korean(text):
    # í•œê¸€, ê³µë°±, ê¸°ë³¸ ë¬¸ì¥ë¶€í˜¸ë§Œ í—ˆìš©
    return re.sub(r"[^ê°€-í£0-9\s\.\,\?\!\:\n]", "", text)


def convert_logic(conversation):
    prompt = f"""<|begin_of_text|><|start_header_id|>system<|end_header_id|>
ë„ˆëŠ” í‚¤ì›Œë“œ ì¶”ì¶œ ì „ìš© ëª¨ë¸ì´ë‹¤.

[ì¶œë ¥ ê·œì¹™]
- ë°˜ë“œì‹œ í•œê¸€ ëª…ì‚¬ 3ê°œë§Œ ì¶œë ¥í•œë‹¤.
- ê° ë‹¨ì–´ëŠ” 10ê¸€ì ì´ë‚´ì´ë‹¤.
- ê° ë‹¨ì–´ëŠ” ì¤„ë°”ê¿ˆìœ¼ë¡œ êµ¬ë¶„í•œë‹¤.
- ë¬¸ì¥, ì¡°ì‚¬, ë™ì‚¬, í˜•ìš©ì‚¬ ì‚¬ìš© ê¸ˆì§€.
- ì„¤ëª…, ê¸°í˜¸, ë²ˆí˜¸, ë§ˆì¹¨í‘œ ì ˆëŒ€ ê¸ˆì§€.
- 3ì¤„ì´ ì•„ë‹ˆë©´ ì˜¤ë‹µì´ë‹¤.

ì˜ˆì‹œ ì¶œë ¥ í˜•ì‹:
ì‚¬ê³¼
í•™êµ
ì—¬í–‰

<|eot_id|>
<|start_header_id|>user<|end_header_id|>
ë‹¤ìŒ ëŒ€í™”ì˜ í•µì‹¬ ì†Œì¬ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ëª…ì‚¬ 3ê°œë§Œ ì¶œë ¥í•˜ë¼.

{conversation}

<|eot_id|>
<|start_header_id|>assistant<|end_header_id|>
"""
    try:
        result = pipe(
            prompt,
            max_new_tokens=50,
            do_sample=True,
            temperature=0.2, #ì°½ì˜ì„±
            top_p=0.9,
            repetition_penalty=1.2,
            return_full_text=False,
            eos_token_id=tokenizer.eos_token_id
        )

        output = result[0]['generated_text'].strip()
        output = keep_only_korean(output)

        # ëª¨ë¸ì´ 'ë³€í™˜:' ì´ë¼ëŠ” ë‹¨ì–´ê¹Œì§€ ê°™ì´ ì¶œë ¥í•  ìˆ˜ ìˆìœ¼ë¯€ë¡œ ì œê±°
        clean_output = output.replace("ë³€í™˜:", "").strip()
        # í˜¹ì‹œ ëª¨ë¥¼ ì‹œìŠ¤í…œ íƒœê·¸ ì œê±°
        clean_output = clean_output.split("<|eot_id|>")[0].strip()
        clean_output = clean_output.split("[Rewritten Dialogue]")[0].strip()

        # 4. íŠ¹ìˆ˜ë¬¸ì ì •ì œ (í•œê¸€, ìˆ«ì, ëŒ€í™” ê¸°í˜¸ì¸ ì½œë¡ ê³¼ ê¸°ë³¸ì ì¸ ë¶€í˜¸ë§Œ ìœ ì§€)
        clean_output = re.sub(r"[^ê°€-í£0-9\s\.\,\?\!\n]", "", clean_output)

        return clean_output

    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
# 4ï¸âƒ£ ë³€í™˜ í•¨ìˆ˜ (ìµœì¢… ìˆ˜ì • ë²„ì „)
import re

def keep_only_korean(text):
    # í•œê¸€, ê³µë°±, ê¸°ë³¸ ë¬¸ì¥ë¶€í˜¸ë§Œ í—ˆìš©
    return re.sub(r"[^ê°€-í£0-9\s\.\,\?\!\:\n]", "", text)


def convert_persona_logic(conversation):
    prompt = f"""<|begin_of_text|><|start_header_id|>system<|end_header_id|>
ë„ˆëŠ” ëŒ€í™” ë¶„ì„ ì „ë¬¸ê°€ì´ë‹¤.
ë‹¤ìŒ ëŒ€í™”ë¥¼ ë¶„ì„í•˜ì—¬ ì•„ë˜ 3ê°€ì§€ ì •ë³´ë¥¼ ê°ê° í•œ ì¤„ì”© ì¶œë ¥í•˜ë¼.

1. ì§ì—…
2. ê´€ê³„
3. ì¥ì†Œ

- ì„¤ëª… ê¸ˆì§€
- ì¶”ê°€ ë¬¸ì¥ ê¸ˆì§€
- ë°˜ë“œì‹œ 3ì¤„ë§Œ ì¶œë ¥

<|eot_id|>
<|start_header_id|>user<|end_header_id|>
{conversation}
<|eot_id|>
<|start_header_id|>assistant to=pipe.generate
"""
    try:
        result = pipe(
            prompt,
            max_new_tokens=50,
            do_sample=True,
            temperature=0.2, #ì°½ì˜ì„±
            top_p=0.9,
            repetition_penalty=1.2,
            return_full_text=False,
            eos_token_id=tokenizer.eos_token_id
        )

        output = result[0]['generated_text'].strip()
        output = keep_only_korean(output)

        # ëª¨ë¸ì´ 'ë³€í™˜:' ì´ë¼ëŠ” ë‹¨ì–´ê¹Œì§€ ê°™ì´ ì¶œë ¥í•  ìˆ˜ ìˆìœ¼ë¯€ë¡œ ì œê±°
        clean_output = output.replace("ë³€í™˜:", "").strip()
        # í˜¹ì‹œ ëª¨ë¥¼ ì‹œìŠ¤í…œ íƒœê·¸ ì œê±°
        clean_output = clean_output.split("<|eot_id|>")[0].strip()
        clean_output = clean_output.split("[Rewritten Dialogue]")[0].strip()

        # 4. íŠ¹ìˆ˜ë¬¸ì ì •ì œ (í•œê¸€, ìˆ«ì, ëŒ€í™” ê¸°í˜¸ì¸ ì½œë¡ ê³¼ ê¸°ë³¸ì ì¸ ë¶€í˜¸ë§Œ ìœ ì§€)
        clean_output = re.sub(r"[^ê°€-í£0-9\s\.\,\?\!\n]", "", clean_output)
        # ì •ê·œì‹ìœ¼ë¡œ "1. ì§ì—…", "2. ê´€ê³„", "3. ì¥ì†Œ" ì œê±°
        clean_output = re.sub(r"^\d+\.\s*(ì§ì—…|ê´€ê³„|ì¥ì†Œ)\s*", "", clean_output, flags=re.MULTILINE)

        return clean_output

    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
import re
import random
import gc
import torch
import pandas as pd

def convert_general_conversation_logic(persona, keywords):

  PROMPT = """
  ë‹¹ì‹ ì€ í•œêµ­ì¸ë“¤ì˜ ì‹¤ì œ ì¼ìƒ ëŒ€í™”ë¥¼ ì™„ë²½í•˜ê²Œ ì¬í˜„í•˜ëŠ” ì‹œë‚˜ë¦¬ì˜¤ ì‘ê°€ì…ë‹ˆë‹¤.
  """

  instruction = f"""
  [ì—­í•  ì„¤ì •] {persona}
  [ì£¼ì œ í‚¤ì›Œë“œ] {keywords}

  [ì§€ì‹œ ì‚¬í•­]
  1. ìœ„ í‚¤ì›Œë“œë¥¼ ì‚¬ìš©í•´ í•œêµ­ì¸ ë‘ ëª…ì˜ ìì—°ìŠ¤ëŸ¬ìš´ ìˆ˜ë‹¤ë¥¼ ì‘ì„±í•˜ì„¸ìš”.
  2. ë¬¸ì¥ì€ ë°˜ë“œì‹œ ë§ˆì¹¨í‘œ(.)ë‚˜ ë¬¼ìŒí‘œ(?)ë¡œ ëë§ºìŒë˜ì–´ì•¼ í•©ë‹ˆë‹¤.
  3. í™”ì í‘œê¸°(A:, B:)ë‚˜ ë¶€ì—° ì„¤ëª…ì€ ì ˆëŒ€ í•˜ì§€ ë§ˆì„¸ìš”.
  4. ì¤‘ê°„ì— ëŠê¸°ê±°ë‚˜ ê¹¨ì§„ ë‹¨ì–´ë¥¼ ì ˆëŒ€ ì‚¬ìš©í•˜ì§€ ë§ˆì„¸ìš”.

  [ëŒ€í™” ì‹œì‘]
  """

  messages = [
      {"role": "system", "content": PROMPT},
      {"role": "user","content": instruction}
  ]

  prompt = tokenizer.apply_chat_template(
      messages,
      tokenize=False,
      add_generation_prompt=True
  )

  # ğŸ”¥ assistant, ì‹œìŠ¤í…œ í† í° ì¶œë ¥ ë°©ì§€
  bad_words = ["assistant", "<|start_header_id|>", "<|end_header_id|>"]
  bad_words_ids = [tokenizer(word, add_special_tokens=False).input_ids for word in bad_words]

  result = pipe(
      prompt,
      max_new_tokens=300,
      max_length=None,
      do_sample=True,
      temperature=0.5,          # ğŸ”¥ ì•ˆì •í™”
      top_p=0.9,               # ğŸ”¥ ë…¸ì´ì¦ˆ ê°ì†Œ
      repetition_penalty=1.2,  # ğŸ”¥ ë°˜ë³µ ë°©ì§€
      no_repeat_ngram_size=3, # 3ê°œ ì´ìƒì˜ í† í°ì´ ë°˜ë³µë˜ëŠ” ê²ƒì„ ë¬¼ë¦¬ì ìœ¼ë¡œ ë§‰ìŒ
      bad_words_ids=bad_words_ids, # ìƒí•œ ë¬¸ì(ë² íŠ¸ë‚¨ì–´ ë“±) ì–µì œ
      eos_token_id=tokenizer.eos_token_id,
      return_full_text=False,
  )

  output = result[0]["generated_text"].strip()
  output = re.sub(r'^(A|B)[:ï¼š]\s*', '', output, flags=re.MULTILINE) # í™”ì í‘œì‹œ ì œê±°
  output = re.sub(r'[^ê°€-í£0-9\s\.\,\!\?]', '', output) # ë¶ˆí•„ìš” ê³µë°± ì •ë¦¬
  output = re.sub(r'\n\s*\n', '\n', output).strip()

  return output
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import numpy as np
import re
from tqdm import tqdm
import gc

# 1ï¸âƒ£ ëª¨ë¸ ë° ì„¤ì • ë¡œë“œ
print("ğŸ“¦ ëª¨ë¸ ë° ë°ì´í„° ë¡œë”© ì¤‘...")
lm_model_name = "skt/kogpt2-base-v2"
tokenizer = AutoTokenizer.from_pretrained(lm_model_name)
lm_model = AutoModelForCausalLM.from_pretrained(lm_model_name).to('cuda')
lm_model.eval()

embed_model = SentenceTransformer('jhgan/ko-sroberta-multitask').to('cuda')

# 2ï¸âƒ£ í’ˆì§ˆ í‰ê°€ í•¨ìˆ˜ ì •ì˜
def compute_perplexity(text):
    if not isinstance(text, str) or len(text) < 5: return 999.0
    encodings = tokenizer(text, return_tensors='pt').to('cuda')
    input_ids = encodings.input_ids
    with torch.no_grad():
        outputs = lm_model(input_ids, labels=input_ids)
        loss = outputs.loss
    return torch.exp(loss).item()

# 3ï¸âƒ£ ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸° ë° ì§€í‘œ ê³„ì‚°
df = pd.read_csv("drive/MyDrive/Colab Notebooks/DLthon/train_with_generated_conv.csv")

print("ğŸ“Š í’ˆì§ˆ ì§€í‘œ ê³„ì‚° ì‹œì‘ (Embedding & Perplexity)...")
# ì„ë² ë”© ê³„ì‚°
embeddings = embed_model.encode(df['generated_conversation'].fillna("").tolist(), convert_to_tensor=True)
reference_texts = [
    "ì•ˆë…•í•˜ì„¸ìš”. ì˜¤ëŠ˜ ë‚ ì”¨ ì¢‹ë„¤ìš”.", "ê·¸ë¬êµ¬ë‚˜, ì •ë§ í˜ë“¤ì—ˆê² ë‹¤.",
    "ë‹¤ìŒ ì£¼ë§ì— ì‹œê°„ ë˜ë©´ ê°™ì´ ê°ˆë˜?", "ì§„ì§œ? ëŒ€ë°•ì´ë‹¤! ê·¸ë˜ì„œ ì–´ë–»ê²Œ ëì–´?"
]
ref_embeddings = embed_model.encode(reference_texts, convert_to_tensor=True)
df['embedding_score'] = util.cos_sim(embeddings, ref_embeddings).mean(dim=1).cpu().numpy()

# ì „ì²´ ë°ì´í„° PPL ê³„ì‚° (ì •í™•í•œ í•„í„°ë§ì„ ìœ„í•´ ì „ì²´ ìˆ˜í–‰ ê¶Œì¥)
tqdm.pandas()
df['perplexity'] = df['generated_conversation'].progress_apply(compute_perplexity)

# 4ï¸âƒ£ í•„í„°ë§ ë° ì¬ìƒì„± ëŒ€ìƒ ë¶„ë¥˜
PPL_THRESHOLD = 150.0
SIMILARITY_THRESHOLD = 0.3

def is_garbage(text):
    if not isinstance(text, str) or len(text) < 30: return True
    if re.search(r'[a-zA-Z]', text): return True # ì˜ì–´ í¬í•¨
    if len(set(text.split())) < 8: return True # ë°˜ë³µ ë‹¨ì–´ ê³¼ë‹¤
    return False

# ì¬ìƒì„± ëŒ€ìƒ ì¸ë±ìŠ¤ ì¶”ì¶œ
mask_bad = (df['perplexity'] >= PPL_THRESHOLD) | \
           (df['embedding_score'] <= SIMILARITY_THRESHOLD) | \
           (df['generated_conversation'].apply(is_garbage))

to_regenerate = df[mask_bad].copy()
df_good = df[~mask_bad].copy()

print(f"âœ… ì–‘í˜¸í•œ ë°ì´í„°: {len(df_good)}ê°œ")
print(f"ğŸ”„ ì¬ìƒì„± í•„ìš” ë°ì´í„°: {len(to_regenerate)}ê°œ")

# 5ï¸âƒ£ ë°ì´í„° ì¬ìƒì„± ë° ì—…ë°ì´íŠ¸
if len(to_regenerate) > 0:
    print(f"âœ¨ í’ˆì§ˆ ë¯¸ë‹¬ ë°ì´í„° {len(to_regenerate)}ê°œë¥¼ ì¬ìƒì„±í•©ë‹ˆë‹¤...")
    new_convs = []

    for idx, row in tqdm(to_regenerate.iterrows(), total=len(to_regenerate)):
        try:
            # ì´ì „ì— ì •ì˜í•œ ìƒì„± í•¨ìˆ˜ í˜¸ì¶œ
            new_text = convert_general_conversation_logic(row['persona'], row['keywords'])
            new_convs.append(new_text)
        except:
            new_convs.append(row['generated_conversation']) # ì‹¤íŒ¨ì‹œ ìœ ì§€

    to_regenerate['generated_conversation'] = new_convs

    # ì›ë³¸ DF ì—…ë°ì´íŠ¸
    df.update(to_regenerate)

    # ë©”ëª¨ë¦¬ ì •ë¦¬
    gc.collect()
    torch.cuda.empty_cache()

# 6ï¸âƒ£ ìµœì¢… ì •ì œ ë° ì €ì¥
def final_clean(text):
    text = re.sub(r'[^ê°€-í£\s\.\?\!\,\~]', '', str(text))
    return text.strip()

df['generated_conversation'] = df['generated_conversation'].apply(final_clean)
# ì¤‘ë³µ ì œê±°
df = df.drop_duplicates(subset=['generated_conversation'], keep='first')

save_path = "drive/MyDrive/Colab Notebooks/DLthon/final_refined_data.csv"
df.to_csv(save_path, index=False, encoding='utf-8-sig')

print(f"ğŸ ëª¨ë“  ê³µì •ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤! ìµœì¢… ì €ì¥: {save_path}")
```
